```python

import pandas as pd
df=pd.read_csv("diabetes_Original_Data.csv")
m=0
diabetic=0
limit1=2.0
limit2 =2.42
for index, rowData in df.iterrows():
    if(rowData["DiabetesPedigreeFunction"]>limit1 and rowData["DiabetesPedigreeFunction"]<=limit2):
        m = m+ 1
        if rowData["Outcome"]==1:
          diabetic+=1

print('Total  Count-->',m)
print('Daibitic  Count-->',diabetic)
print('Non-Daibitic  Count-->',m-diabetic)
```

```python
import pandas as pd
import matplotlib.pyplot as plt
import numpy as np

# Read the CSV file
df = pd.read_csv("Diabetes_ori.csv")

x = np.arange(len(df["DiabetesPedigreeFunction"]))
width = 0.35

fig, ax = plt.subplots()

# Create bars
bars1 = ax.bar(x - width/2, df["Count"], width, label="Total Count")
bars2 = ax.bar(x + width/2, df["Diabetic"], width, label="Diabetic")

# Add labels
ax.set_xlabel("DiabetesPedigreeFunction")
ax.set_ylabel("Count")
ax.set_title("Diabetes Pedigree Function vs Count")
ax.set_xticks(x)
ax.set_xticklabels(df["DiabetesPedigreeFunction"])
ax.legend()
ax.bar_label(bars1, padding=3)
ax.bar_label(bars2, padding=3)

plt.show()
```
