```python
import pandas as pd
import matplotlib.pyplot as plt
from sklearn.linear_model import LinearRegression

df = pd.read_csv("Diabetes_ori.csv")

df["DiabetesPedigreeFunction"] = (
    df["DiabetesPedigreeFunction"]
    .str.split("-")
    .apply(lambda x: (float(x[0]) + float(x[1])) / 2)
)

# Remove missing values
df=df.dropna(subset=["DiabetesPedigreeFunction", "percentage"])

# Define X and y correctly
X = df[["DiabetesPedigreeFunction"]]   # 2D
y = df["percentage"]

# Create model
model = LinearRegression()

# Fit model
model.fit(X, y)

print("Model trained successfully")
value = 2.4  # example DPF value

new_data = pd.DataFrame({
    "DiabetesPedigreeFunction": [value]
})

predicted_percentage = model.predict(new_data)

print("Predicted Percentage:", predicted_percentage[0])
predictions = model.predict(X)
error = mean_absolute_error(y, predictions)

print("Mean Absolute Error:", error)
plt.scatter(X, y, label="Actual Data")

plt.plot(X, model.predict(X), color="red", label="Regression Line")

plt.xlabel("DiabetesPedigreeFunction")
plt.ylabel("Percentage")
plt.title("DiabetesPedigreeFunction vs Percentage")

plt.legend()
plt.show()
```
